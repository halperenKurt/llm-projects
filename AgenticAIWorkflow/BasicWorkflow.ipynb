```python
import ollama
import os 

ollama_api=os.getenv("OLLAMA_API")
MODEL = "llama3.2"

messages = [{"role":"user","content":"pick a business area that might be worth exploring fo an Agentic AI oppurtunity"}]

response = ollama.chat(
    model = MODEL,
    messages = messages
)

print(response["message"]["content"])

business_idea = response["message"]["content"]
```

```python
messages = [{"role":"user","content":business_idea + "present a pain-point in that industry - something challenging that might be ripe for an Agentic solution"}]

response = ollama.chat(
    model = MODEL,
    messages = messages
)

print(response["message"]["content"])
```
